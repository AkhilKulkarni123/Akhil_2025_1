Week 1: Initial Setup and Level Creation

Added the Magic and City levels to the assets and navigation folders in the repository.

Created the structure for both levels with initial design concepts.

No coding was done yet, focusing purely on the game world design and level creation.


Week 2: Setting Up the Environment and Basic Gameplay Mechanics

Modified GameEnv.js for the the rendering of the new levels which manages the gameâ€™s environmental properties.

Edited GameControl.js to add the new level file names to the list of rendered levels in the game
This week established the foundational mechanics that allow for basic gameplay.

Mainly Ideation Week with finalizing logic and design process for the new levels logic annd implementation for the new few weeks.


Week 3: Implementing the City Level

Developed GameLevelCity.js to define the gameplay for the City level.

Integrated the following features:

Background with an image of a city map.

New playable character: Stock Guy, with movement using W, A, S, D keys.

Implemented collision detection between the player and NPCs.

Added a collision counter that updates when the player collides with NPCs.

Created an ESC key event conditional to allow to go to the new level or not to check if enough collisions occurred to complete the level.

Impact: The city level became fully interactive, allowing player movement, NPC interaction, and progression tracking, while also monitoring if the user beat the level or not and was allowed to progress based on the logic of them colliding with the NPC's in the new level.


Week 4: Implementing the Magic Level

Developed GameLevelMagic.js to define the gameplay for the Magic level.

Introduced:

Background featuring a castle in the desert.

New playable character: Magic Character, a castle wanderer.

Player movement and animation with sprite sheet support.

Level objects and interactions with the environment.

Quiz was added with questions being mapped to the correct and allowing inut, checking all answers and checking if they were right or not to allow further progression, added a real goal to the game
Restart feature using location.reload was added, restarting the level if not all answers were correct.

Impact: The Magic level was now playable, expanding the game world and providing new challenges.


Week 5: Refining Game Mechanics and Enhancing Features (Tuning)

Improved collision detection for better accuracy in city level.

Implemented a countdown timer (60 seconds) to create urgency in completing levels.

Adjusted sprite scaling and hitbox sizes for better visual representation.

Added UI elements (collision counter display) to improve player feedback and input, like the interactice quiz in the 4th Magic Level

Impact: These refinements made the game more polished, interactive, and engaging.

By the end of Week 5, the game had two fully functional levels, a solid foundation for expansion, and enhanced mechanics for an improved player experience.
