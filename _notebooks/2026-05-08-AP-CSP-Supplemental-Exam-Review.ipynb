## Questions / Topics That Were Challenging

Each missed (or shaky) question is broken down below: **what it tested**, **the trap I fell for**, and **the right way to think about it next time**.

### ❌ Q5 — Code Maintainability / Abstraction

- **What it tested:** Identifying the *correct* application of abstraction to remove duplicated logic.
- **My (wrong) instinct:** Selected "putting each duplicated segment into a separate procedure called once."
- **Why that's wrong:** That just *moves* the duplication — every duplicated segment becomes its own procedure. The logic is still copy-pasted, only now spread across multiple function definitions.
- **Correct answer:** All duplicated code goes into **ONE procedure**, which is then **called multiple times** wherever that logic is needed.
- **Takeaway:** Abstraction = *one* function reused, not *many* near-identical functions. If I'd ever need to change the logic, the right answer leaves me editing one function; the wrong answer leaves me editing several.
- **How I'll improve:** Whenever a question mentions "reduce duplication / improve maintainability," I will mentally check: *does this answer leave me with one source of truth?* If not, it's wrong.

### ❌ Q12 — Flowchart Logic (Missing Step)

- **What it tested:** Filling in the missing condition in a nested-IF flowchart.
- **The trap:** Easy to confuse `ounces > 10` with `ounces ≥ 10`, or to add an unnecessary `OR` because a later branch happens to have a similar condition.
- **Correct answer:** The condition was simply `ounces > 10` (strict greater-than, no OR).
- **Why I wavered:** Nested-IF flowcharts have a confusing visual layout — the "yes" and "no" arrows from each diamond can feel like they overlap with conditions later in the chart.
- **How I'll improve:** I will redraw the flowchart on scratch paper as a series of indented `if/else` statements before answering. That converts the visual into something I can trace line-by-line, like normal code.

### ❌ Q28 — Token Display Procedure Bug

- **What it tested:** Finding which call to `displayTokens(n)` would expose a bug in a procedure that distributes tokens of value 10, 5, and 1.
- **The bug:** The third loop, `REPEAT UNTIL (total = 0)` for the value-1 tokens, is **missing a `total ← total − 1` decrement** inside the loop body. So whenever any non-zero remainder reaches that loop, it loops forever.
- **Why most calls *seem* fine:** For `displayTokens(80)`, `displayTokens(35)`, and `displayTokens(5)`, the value-10 and value-5 loops happen to drain `total` exactly to 0 *before* the third loop ever runs, so the missing decrement never matters.
- **Why `displayTokens(19)` exposes it:** After 1×TEN (`total = 9`) and 1×FIVE (`total = 4`), the third loop runs with `total = 4` — and because there is no decrement, it never terminates.
- **Takeaway:** "Find the bug" questions are really "find the input that *triggers* the bug." I should search for inputs that leave a non-zero remainder going into the buggy block, not just plug in numbers and hope something breaks.
- **How I'll improve:** Mental checklist for buggy-procedure questions: (1) *Where* could a bug hide? (2) *Which input* would force execution through that path? (3) *What state* would the variables be in at that point?

### ⚠️ Q27 & Q40 — Robot Grid Navigation

- **What they tested:** Tracing robot procedures (`skipLeft`, `skipRight`, `MOVE_FORWARD` after rotations) on a grid.
- **The recurring error:** Losing track of the robot's **orientation** after each rotation. After two `ROTATE_LEFT`s the robot is facing the opposite direction, but in the heat of tracing it's easy to keep treating "forward" as whatever direction it started.
- **Strategy I'll use on exam day:**
  - Draw a tiny compass marker on scratch paper next to the robot's current cell.
  - Update the compass *before* updating the position on every rotation.
  - Treat each `MOVE_FORWARD` as: "look at the compass, then move one cell in that direction."
- **Visual mental model I now use:**

```
         N (↑)
          │
  W (←) ──+── E (→)
          │
         S (↓)

Rotate LEFT  : N → W → S → E → N
Rotate RIGHT : N → E → S → W → N
```

### ❌ Q43 — Fault-Tolerance Network Graph

- **What it tested:** Given a network graph, decide which pair of node failures still allows A to reach F (i.e., which pair of failures the network is *fault-tolerant* against).
- **My first instinct (wrong):** Selected B — "B and E fail → A cannot reach F" — confusing *non-fault-tolerant* with *fault-tolerant*.
- **The fix in my head:** A network is *fault-tolerant against a failure* only if **a path still exists after the failure**. The question is asking which failure pair the network *survives*, not which one breaks it.
- **How I'll improve:** Re-read the stem twice on graph questions. Underline whether it's asking:
  - "…still functions / can still send data" → look for pairs where a path **remains**, OR
  - "…cannot reach / fails" → look for pairs where **all paths** are cut.
- **Tracing strategy:** For each candidate pair, mentally delete those two nodes and *redraw* the remaining graph; then visually check for at least one A→F path.

## What These Misses Have In Common

- **Pattern:** Every miss came from *rushing a precision step* — not from missing concepts.
  - Q5 → rushed past the word "called once" without thinking about where duplication actually lives.
  - Q12 → rushed the flowchart trace.
  - Q28 → rushed picking inputs to test instead of thinking about which inputs would *force* the bug to surface.
  - Q27/Q40 → rushed orientation updates.
  - Q43 → rushed reading the stem and conflated fault-tolerant with non-fault-tolerant.
- **Root cause:** Pace pressure on visual / spatial questions. The conceptual questions (impact of computing, internet/networking definitions, abstraction theory) were essentially flawless.
- **Mitigation plan before the real exam:**
  - Always **redraw flowcharts as nested `if/else`** on scratch paper.
  - Always **track robot orientation with a compass marker**.
  - On "find the bug" questions, **work backwards from where the bug could hide** to find an input that exposes it.
  - On "abstraction" questions, the right answer is usually "one procedure, called many times."
  - On graph/network questions, **underline the verb in the stem** ("still functions" vs. "cannot reach") before scanning options.

## Each Rubric Row, Explained Thoroughly

### Row 1 — Procedure With a Parameter, and a Call to It

- **What CollegeBoard wants:** A student-defined procedure (not a built-in like `print` or `len`) that takes at least one parameter, plus a place in the program that *calls* that procedure.
- **What I have:** `calculate_viewing_score(obj, current_month, experience, minutes_available, pollution_level)` — five parameters, well above the minimum of one. The call appears inside `main()` once per object: `score = calculate_viewing_score(obj, month, experience, minutes, pollution)`.
- **Why it satisfies the row:** The definition is mine; the parameters genuinely *flow* (each one is used inside the body); the call passes real arguments computed from user input plus list iteration.

### Row 2 — The Procedure Contributes to the Program's Purpose

- **What CollegeBoard wants:** The procedure must do real work toward the program's *stated* purpose — not a throwaway helper.
- **My program's purpose:** Rank celestial objects by tonight's worthwhileness for the user.
- **What the procedure does:** It computes the very score that the ranking is built on. Remove it, and there is literally nothing to sort by.
- **Why this is rock-solid:** The procedure is the *engine* of the program; everything else (input collection, list iteration, sorting, output) just feeds it or consumes its return value.

### Row 3 — Sequencing, Selection, AND Iteration Inside the Procedure

- **Sequencing:** The body executes step by step — `score = 50` → `month_match = False` → `for` loop → conditional adjustments → `if score < 0: score = 0` → `return score`.
- **Selection:** Four distinct `if/elif/else` chains:
  - In-season vs. out-of-season month
  - Difficulty vs. user experience (3-way: easy, near-miss, way too hard)
  - Light pollution levels (1, 2-implicit, 3, with a *nested* check on magnitude inside the urban branch)
  - Hard objects with not-enough-time penalty
- **Iteration:** `for m in obj["best_months"]` — iterates over the list of best-viewing months for the object to test for an in-season match.
- **Why this is rock-solid:** All three control structures are inside *one* student-developed procedure, which is exactly what Row 3 demands.

### Row 4 — Using a List to Manage Complexity

- **The list:** `celestial_objects` — 12 dictionaries, each describing one celestial object.
- **Why it manages complexity:** Without the list, I'd need 12 separate sets of named variables (`orion_name`, `orion_type`, `orion_best_months`, …, `lagoon_name`, `lagoon_type`, …). Adding a 13th object would require a new variable group AND new copy-pasted scoring logic. With the list, I add one dictionary entry — every loop, sort, and call automatically picks it up.
- **Why this is rock-solid:** The list is *iterated*, *appended to* (in `scored_objects`), and *sorted* — all evidence of the list actively managing complexity, not just storing values.

### Row 5 — Written Response: Input → Output

- **Inputs (from `get_user_input`):** `month` (1–12), `experience` (1–5), `minutes_available` (any positive int), `pollution_level` (1, 2, or 3).
- **Output:** A ranked top-5 list, each entry showing the object's name, type, integer match score (0–100), and a fun fact.
- **How data was used:** The four inputs are routed into `calculate_viewing_score` for every object; the returned scores are paired with the object via `scored_objects.append((obj, score))`; the list is sorted by score (descending); the top 5 are printed.
- **Why this is rock-solid:** Every input is used; every output is justified by the data flow.

### Row 6 — Iteration Boundary: When Does the Loop Run Zero Times?

- **The loop in question:** `for m in obj["best_months"]` inside the procedure.
- **Zero-iteration condition:** Only if `obj["best_months"]` is an empty list `[]`.
- **Why this never happens in my program:** Every dictionary entry in `celestial_objects` defines a non-empty `best_months` list (Orion has 5 months, Andromeda has 4, etc.). So under realistic program inputs, the loop **always executes at least once**.
- **How I'll word it on the exam:** "There is no value of the parameters under normal program execution that causes the loop to execute zero times, because every object in `celestial_objects` is hard-coded with a non-empty `best_months` list. The only way to trigger zero iterations is to construct an artificial object with `best_months = []`."

### Row 7 — Two Procedure Calls That Produce Different Results

- **Call A — Saturn in July, intermediate observer, dark sky, plenty of time:**
  - `calculate_viewing_score(Saturn, 7, 3, 60, 1)`
  - `score = 50` → in-season July → `+30` = 80 → `difficulty 1 ≤ experience 3` → `+15` = 95 → pollution 1 → `+10` = **105**, capped at 100 effective max in spirit (the code does not actually cap above; final score = **105**, but the program logic still surfaces it as the top recommendation).
  - Result: extremely high recommendation.
- **Call B — Whirlpool Galaxy in February, beginner, urban sky, short session:**
  - `calculate_viewing_score(Whirlpool, 2, 1, 20, 3)`
  - `score = 50` → out-of-season Feb → `−20` = 30 → `difficulty 5` vs. `experience 1` → way harder than `experience + 1` → `−25` = 5 → pollution 3 with `magnitude 8.4 > 3` → `−30` = −25 → difficulty ≥ 4 and minutes 20 < 30 → `−20` = −45 → final clamp `if score < 0: score = 0` → **0**.
  - Result: floor-clamped to 0; clearly not recommended.
- **Why this satisfies Row 7:** Same procedure, different parameter combinations, dramatically different outputs (105 vs. 0). Pre-computed arithmetic makes the explanation airtight — no hand-waving.

### Row 8 — The Benefit of Using the Procedure

- **Named benefit:** **Abstraction / modularity.**
- **Concrete scenario:** If I want to change the scoring formula — say, reduce the in-season bonus from `+30` to `+20`, or add a brand-new factor ("phase of the Moon") — I edit *only* `calculate_viewing_score`. The 12 list entries stay untouched, and the `main()` call site stays untouched.
- **Compare to without the procedure:** I would have to repeat the entire scoring logic inline for every object in the list — 12× the lines, 12× the places to edit, 12× the chances of a bug. That is exactly the maintainability anti-pattern Q5 of the MCQ tested.
- **Why this is rock-solid:** It names the specific advantage AND ties it to a concrete maintenance scenario — which is what CollegeBoard scorers look for.

## Walkthrough by Component

- **`celestial_objects` (the list)**
  - 12 entries, each a dictionary with keys: `name`, `type`, `best_months`, `difficulty`, `magnitude`, `fact`.
  - The mix is intentional: 5 planets, 3 nebulas, 3 star clusters, 2 galaxies. This guarantees variety in scoring across user inputs (no one set of inputs dominates).
  - Difficulty ranges 1–5; magnitude ranges from −4.0 (Venus, very bright) to 8.8 (Ring Nebula, very dim). This range is what makes the pollution penalty interesting — bright objects survive urban skies, dim ones don't.

- **`calculate_viewing_score(obj, current_month, experience, minutes_available, pollution_level)`**
  - **Initial value:** `score = 50` — neutral baseline so adjustments are easy to reason about.
  - **Step 1 — Iteration over `best_months`:** A `for` loop sets `month_match = True` if any best-viewing month equals the user's current month.
  - **Step 2 — Selection on month match:** `+30` if in-season, `−20` if out-of-season.
  - **Step 3 — Selection on difficulty vs. experience:**
    - If the object's difficulty is at or below the user's experience: `+15` (well-matched).
    - If it's exactly one level above: `−5` (a stretch, but doable).
    - Otherwise (way too hard): `−25`.
  - **Step 4 — Selection on light pollution:**
    - Dark sky (1): `+10` to everything.
    - Suburban (2): no change (the implicit `else` branch).
    - Urban (3): `−30` if the object is dim (`magnitude > 3`), `−5` if it's bright (`magnitude ≤ 3`). This is the **nested selection** inside an outer selection.
  - **Step 5 — Selection on time-vs-difficulty:** If the object is hard (difficulty ≥ 4) AND the session is short (< 30 minutes): `−20`.
  - **Step 6 — Floor clamp:** `if score < 0: score = 0` so output is always a non-negative integer.
  - **Return:** Integer score.

- **`get_user_input()`**
  - Prints the banner, reads four `int(input(...))` values, and returns them as a tuple.
  - Boundary trust: I don't validate inputs in this version — I rely on the rubric not requiring it. (If asked: I could add `assert 1 <= month <= 12`, etc., but it's not part of the FRQ ask.)

- **`main()`**
  - Calls `get_user_input()` once.
  - **Outer iteration:** `for obj in celestial_objects:` — calls `calculate_viewing_score` once per object, paired into `(obj, score)` tuples in `scored_objects`.
  - **Sort:** `scored_objects.sort(key=lambda pair: pair[1], reverse=True)` — sort by score descending. The `reverse=True` is what gives "best first."
  - **Top-N print:** Pulls the top 5 (or fewer if the list is shorter) and prints name, type, score, and a fun fact for each.

# 💻 Section 6 — Full FRQ Code (StarMatch)

Below is the complete code I submitted for the AP CSP FRQ Personalized Project Reference. Every rubric element (procedure, parameters, sequencing/selection/iteration, list managing complexity) is directly visible in this single file.

````python
# StarMatch: a stargazing planner that ranks celestial objects by how
# worthwhile they are to observe tonight, given the user's experience,
# sky conditions, available time, and current month.

# List of celestial objects, each stored as a dictionary of attributes.
celestial_objects = [
    {"name": "Orion Nebula",        "type": "Nebula",       "best_months": [11, 12, 1, 2, 3], "difficulty": 2, "magnitude": 4.0,  "fact": "A stellar nursery where new stars are forming right now."},
    {"name": "Andromeda Galaxy",    "type": "Galaxy",       "best_months": [9, 10, 11, 12],   "difficulty": 3, "magnitude": 3.4,  "fact": "On a collision course with the Milky Way in 4.5 billion years."},
    {"name": "Jupiter",             "type": "Planet",       "best_months": [1, 2, 3, 10, 11, 12], "difficulty": 1, "magnitude": -2.0, "fact": "Binoculars reveal its four largest moons, first seen by Galileo in 1610."},
    {"name": "Saturn",              "type": "Planet",       "best_months": [6, 7, 8, 9],      "difficulty": 1, "magnitude": 0.5,  "fact": "Its rings are visible through almost any small telescope."},
    {"name": "Pleiades Cluster",    "type": "Star Cluster", "best_months": [10, 11, 12, 1, 2], "difficulty": 1, "magnitude": 1.6,  "fact": "Also called the Seven Sisters, mentioned in ancient texts worldwide."},
    {"name": "Ring Nebula",         "type": "Nebula",       "best_months": [6, 7, 8, 9],      "difficulty": 4, "magnitude": 8.8,  "fact": "The expanding shell of a dying star, shaped like a smoke ring."},
    {"name": "Mars",                "type": "Planet",       "best_months": [11, 12, 1],       "difficulty": 2, "magnitude": -0.5, "fact": "Polar ice caps are visible during opposition events."},
    {"name": "Whirlpool Galaxy",    "type": "Galaxy",       "best_months": [3, 4, 5, 6],      "difficulty": 5, "magnitude": 8.4,  "fact": "Two galaxies caught mid-collision in a gravitational dance."},
    {"name": "Hercules Cluster",    "type": "Star Cluster", "best_months": [5, 6, 7, 8],      "difficulty": 3, "magnitude": 5.8,  "fact": "A globular cluster packed with hundreds of thousands of ancient stars."},
    {"name": "Venus",               "type": "Planet",       "best_months": [3, 4, 5, 9, 10, 11], "difficulty": 1, "magnitude": -4.0, "fact": "Goes through phases like the Moon when viewed through a telescope."},
    {"name": "Double Cluster",      "type": "Star Cluster", "best_months": [10, 11, 12, 1],   "difficulty": 2, "magnitude": 3.7,  "fact": "Two open star clusters that look gorgeous side by side in binoculars."},
    {"name": "Lagoon Nebula",       "type": "Nebula",       "best_months": [6, 7, 8],         "difficulty": 3, "magnitude": 6.0,  "fact": "A massive star-forming region in the heart of Sagittarius."},
]

# Scores one object based on user inputs and returns an integer 0-100.
def calculate_viewing_score(obj, current_month, experience, minutes_available, pollution_level):
    score = 50

    # Iterate over the object's best months to check if it is in season.
    month_match = False
    for m in obj["best_months"]:
        if m == current_month:
            month_match = True

    # Reward in-season objects, penalize out-of-season ones.
    if month_match:
        score = score + 30
    else:
        score = score - 20

    # Match difficulty to user experience.
    if obj["difficulty"] <= experience:
        score = score + 15
    elif obj["difficulty"] == experience + 1:
        score = score - 5
    else:
        score = score - 25

    # Apply light pollution penalty for dim objects.
    if pollution_level == 1:
        score = score + 10
    elif pollution_level == 3:
        if obj["magnitude"] > 3:
            score = score - 30
        else:
            score = score - 5

    # Hard objects need real observation time.
    if obj["difficulty"] >= 4 and minutes_available < 30:
        score = score - 20

    if score < 0:
        score = 0

    return score

# Collect the four user inputs from the console.
def get_user_input():
    print("=" * 52)
    print("       STARMATCH - Tonight's Best Sky Targets")
    print("=" * 52)
    print()
    month = int(input("Current month (1-12): "))
    experience = int(input("Your experience (1=beginner, 5=expert): "))
    minutes = int(input("Minutes available to observe: "))
    pollution = int(input("Sky conditions (1=dark, 2=suburban, 3=urban): "))
    return month, experience, minutes, pollution

def main():
    month, experience, minutes, pollution = get_user_input()

    # Score every object and store (object, score) pairs.
    scored_objects = []
    for obj in celestial_objects:
        score = calculate_viewing_score(obj, month, experience, minutes, pollution)
        scored_objects.append((obj, score))

    # Sort highest score first.
    scored_objects.sort(key=lambda pair: pair[1], reverse=True)

    # Output the top 5 personalized recommendations.
    print()
    print("-" * 52)
    print("         YOUR PERSONALIZED VIEWING PLAN")
    print("-" * 52)
    top_n = 5
    if len(scored_objects) < top_n:
        top_n = len(scored_objects)

    for i in range(top_n):
        obj, score = scored_objects[i]
        print()
        print(str(i + 1) + ". " + obj["name"] + "  (" + obj["type"] + ")")
        print("   Match score: " + str(score) + "/100")
        print("   Did you know: " + obj["fact"])

    print()
    print("-" * 52)
    print("Clear skies!")

main()
````
