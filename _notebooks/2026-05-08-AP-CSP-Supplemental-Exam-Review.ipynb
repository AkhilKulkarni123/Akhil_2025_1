## Questions / Topics That Were Challenging

The 3 missed questions all share the same root cause — **rushing visual tracing under pace pressure**. Each is broken down below into:

- 🎯 **What it tested**
- ❌ **The trap I fell for**
- ✅ **The right way to think about it next time**


<div class="qcard">
<h3>Q27 &amp; Q40 — Robot Grid Navigation <span class="badge miss">Missed</span></h3>

- 🎯 **What they tested:** Tracing robot procedures (`skipLeft`, `skipRight`, `MOVE_FORWARD` after rotations) on a grid.
- ❌ **The recurring error:** Losing track of the robot's **orientation** after each rotation.
- 🔄 **Why orientation slips:** After two `ROTATE_LEFT`s the robot faces the opposite direction — easy to forget and keep treating "forward" as the original heading.
- ✅ **Strategy I'll use on exam day:**
  - 🧭 Draw a tiny compass marker on scratch paper next to the robot's current cell.
  - 🔁 Update the compass *before* updating the position on every rotation.
  - 🚶 Treat each `MOVE_FORWARD` as: "look at the compass, then move one cell that direction."

**🧭 Visual mental model I now use:**

```
         N (↑)
          │
  W (←) ──+── E (→)
          │
         S (↓)

Rotate LEFT  : N → W → S → E → N
Rotate RIGHT : N → E → S → W → N
```

</div>


<div class="qcard">
<h3>Q43 — Fault-Tolerance Network Graph <span class="badge miss">Missed</span></h3>

- 🎯 **What it tested:** Given a network graph, decide which pair of node failures the network is *fault-tolerant* against (i.e., still allows A → F).
- ❌ **My first instinct (wrong):** Selected B — "B and E fail → A cannot reach F" — confusing *non-fault-tolerant* with *fault-tolerant*.
- 🧠 **The correct frame:** A network is *fault-tolerant against a failure* only if **a path still exists after the failure** — the question asks which pair the network *survives*, not which one breaks it.
- ✅ **How I'll improve — re-read the stem twice:**
  - "…still functions / can still send data" → look for pairs where a path **remains**.
  - "…cannot reach / fails" → look for pairs where **all paths** are cut.
- 🔍 **Tracing strategy:**
  - For each candidate pair, mentally **delete those two nodes**.
  - Redraw the remaining graph.
  - Visually check for at least one A → F path.

</div>

- 🔁 Mentally delete failed nodes and **redraw the graph** before answering fault-tolerance questions.

````python
# StarMatch: a stargazing planner that ranks celestial objects by how
# worthwhile they are to observe tonight, given the user's experience,
# sky conditions, available time, and current month.

# List of celestial objects, each stored as a dictionary of attributes.
celestial_objects = [
    {"name": "Orion Nebula",        "type": "Nebula",       "best_months": [11, 12, 1, 2, 3], "difficulty": 2, "magnitude": 4.0,  "fact": "A stellar nursery where new stars are forming right now."},
    {"name": "Andromeda Galaxy",    "type": "Galaxy",       "best_months": [9, 10, 11, 12],   "difficulty": 3, "magnitude": 3.4,  "fact": "On a collision course with the Milky Way in 4.5 billion years."},
    {"name": "Jupiter",             "type": "Planet",       "best_months": [1, 2, 3, 10, 11, 12], "difficulty": 1, "magnitude": -2.0, "fact": "Binoculars reveal its four largest moons, first seen by Galileo in 1610."},
    {"name": "Saturn",              "type": "Planet",       "best_months": [6, 7, 8, 9],      "difficulty": 1, "magnitude": 0.5,  "fact": "Its rings are visible through almost any small telescope."},
    {"name": "Pleiades Cluster",    "type": "Star Cluster", "best_months": [10, 11, 12, 1, 2], "difficulty": 1, "magnitude": 1.6,  "fact": "Also called the Seven Sisters, mentioned in ancient texts worldwide."},
    {"name": "Ring Nebula",         "type": "Nebula",       "best_months": [6, 7, 8, 9],      "difficulty": 4, "magnitude": 8.8,  "fact": "The expanding shell of a dying star, shaped like a smoke ring."},
    {"name": "Mars",                "type": "Planet",       "best_months": [11, 12, 1],       "difficulty": 2, "magnitude": -0.5, "fact": "Polar ice caps are visible during opposition events."},
    {"name": "Whirlpool Galaxy",    "type": "Galaxy",       "best_months": [3, 4, 5, 6],      "difficulty": 5, "magnitude": 8.4,  "fact": "Two galaxies caught mid-collision in a gravitational dance."},
    {"name": "Hercules Cluster",    "type": "Star Cluster", "best_months": [5, 6, 7, 8],      "difficulty": 3, "magnitude": 5.8,  "fact": "A globular cluster packed with hundreds of thousands of ancient stars."},
    {"name": "Venus",               "type": "Planet",       "best_months": [3, 4, 5, 9, 10, 11], "difficulty": 1, "magnitude": -4.0, "fact": "Goes through phases like the Moon when viewed through a telescope."},
    {"name": "Double Cluster",      "type": "Star Cluster", "best_months": [10, 11, 12, 1],   "difficulty": 2, "magnitude": 3.7,  "fact": "Two open star clusters that look gorgeous side by side in binoculars."},
    {"name": "Lagoon Nebula",       "type": "Nebula",       "best_months": [6, 7, 8],         "difficulty": 3, "magnitude": 6.0,  "fact": "A massive star-forming region in the heart of Sagittarius."},
]

# Scores one object based on user inputs and returns an integer 0-100.
def calculate_viewing_score(obj, current_month, experience, minutes_available, pollution_level):
    score = 50

    # Iterate over the object's best months to check if it is in season.
    month_match = False
    for m in obj["best_months"]:
        if m == current_month:
            month_match = True

    # Reward in-season objects, penalize out-of-season ones.
    if month_match:
        score = score + 30
    else:
        score = score - 20

    # Match difficulty to user experience.
    if obj["difficulty"] <= experience:
        score = score + 15
    elif obj["difficulty"] == experience + 1:
        score = score - 5
    else:
        score = score - 25

    # Apply light pollution penalty for dim objects.
    if pollution_level == 1:
        score = score + 10
    elif pollution_level == 3:
        if obj["magnitude"] > 3:
            score = score - 30
        else:
            score = score - 5

    # Hard objects need real observation time.
    if obj["difficulty"] >= 4 and minutes_available < 30:
        score = score - 20

    if score < 0:
        score = 0

    return score

# Collect the four user inputs from the console.
def get_user_input():
    print("=" * 52)
    print("       STARMATCH - Tonight's Best Sky Targets")
    print("=" * 52)
    print()
    month = int(input("Current month (1-12): "))
    experience = int(input("Your experience (1=beginner, 5=expert): "))
    minutes = int(input("Minutes available to observe: "))
    pollution = int(input("Sky conditions (1=dark, 2=suburban, 3=urban): "))
    return month, experience, minutes, pollution

def main():
    month, experience, minutes, pollution = get_user_input()

    # Score every object and store (object, score) pairs.
    scored_objects = []
    for obj in celestial_objects:
        score = calculate_viewing_score(obj, month, experience, minutes, pollution)
        scored_objects.append((obj, score))

    # Sort highest score first.
    scored_objects.sort(key=lambda pair: pair[1], reverse=True)

    # Output the top 5 personalized recommendations.
    print()
    print("-" * 52)
    print("         YOUR PERSONALIZED VIEWING PLAN")
    print("-" * 52)
    top_n = 5
    if len(scored_objects) < top_n:
        top_n = len(scored_objects)

    for i in range(top_n):
        obj, score = scored_objects[i]
        print()
        print(str(i + 1) + ". " + obj["name"] + "  (" + obj["type"] + ")")
        print("   Match score: " + str(score) + "/100")
        print("   Did you know: " + obj["fact"])

    print()
    print("-" * 52)
    print("Clear skies!")

main()
````
